```python
df_final['Hospital_type_code']=le.fit_transform(df_final['Hospital_type_code'])
df_final['Hospital_region_code']=le.fit_transform(df_final['Hospital_region_code'])
df_final['Department']=le.fit_transform(df_final['Department'])
df_final['Ward_Type']=le.fit_transform(df_final['Ward_Type'])
df_final['Ward_Facility_Code']=le.fit_transform(df_final['Ward_Facility_Code'])
df_final['Type of Admission']=le.fit_transform(df_final['Type of Admission'])
df_final['Severity of Illness']=le.fit_transform(df_final['Severity of Illness'])
df_final['Age'] = le.fit_transform(df_final['Age'])
```

```python
x=df_train_final
x=x.drop(['case_id'],axis=1)
x = x.drop(['is_train'], axis=1)
x = x.drop(['Stay'], axis=1)
```

```python
y=df_train['Stay']
x_pred = df_test_final
x_pred = x_pred.drop(['case_id'], axis=1)
x_pred = x_pred.drop(['is_train'], axis=1)
x_pred = x_pred.drop(['Stay'], axis=1)
```

```python
x.head()
```

```python
params = {}
params['learning_rate'] = 0.09
params['max_depth'] = 50
params['n_estimators'] = 500
params['objective'] = 'multiclass'
params['boosting_type'] = 'gbdt'
params['subsample'] = 0.7
params['random_state'] = 50
params['colsample_bytree']=0.7
params['min_data_in_leaf'] = 55
params['reg_alpha'] = 1.6
params['reg_lambda'] = 1.1
```

```python
clf = lgb.LGBMClassifier(**params)
clf.fit(x, np.ravel(y), eval_metric='multi_error', verbose=False)
predparams= clf.predict(x)
predparams
```
